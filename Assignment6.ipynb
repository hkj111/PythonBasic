## 1. What are escape characters, and how do you use them?
<br>
Escape Characters in pyhton are used to insert characters or white spaces that are illegal in strings or will throw error if they are used without escape charaters. "\" is a escape character. For example if we want to include double quote in a string. s = "Hello my name is "Harsh" " Now this will throw error we can correct it by using "\" like this. s = "Hello My name is \"Harsh\" ".
